```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import pickle
from os import listdir
from os.path import isfile, join
```

```python
header = ["fold", "n_targets", "threshold", "precision", "recall", "f1"]

def f1(precision, recall):
    return 2 * (precision * recall) / (precision + recall)
```

```python
path = "subreddit_topical/"
files = [f for f in listdir(path) if isfile(join(path, f))]

df = pd.DataFrame(columns=header)
for f in files:
    with open(path + f) as file:
        fold = f[5:-5]
        content = [x.strip() for x in file.readlines()]
        for i in range(int(len(content) / 4)):
            tar, th = content[i * 4].split(" ")
            prec = float(content[i * 4 + 2].split("\t")[-1])
            rec = float(content[i * 4 + 3].split("\t")[-1])
            f1_score = f1(prec, rec)
            df = df.append({
             "fold": fold,
             "n_targets": int(tar),
             "threshold": float(th) * 100,
             "precision": prec * 100,
             "recall": rec * 100,
             "f1": f1_score * 100
              }, ignore_index=True)
```

```python
sns.lineplot(data=df, x="threshold", y="precision", label="Precision")
sns.lineplot(data=df, x="threshold", y="recall", label="Recall")
sns.lineplot(data=df, x="threshold", y="f1", label="F1")
plt.ylabel("Model Quality [in %]")
plt.xlabel("Prediction Theshold [in %]")
plt.savefig(path[:-1] + "_threshold.pdf")
```

```python
tmp = df.loc[df['threshold'] == 20.0]

sns.lineplot(data=tmp, x="n_targets", y="precision", label="Precision")
sns.lineplot(data=tmp, x="n_targets", y="recall", label="Recall")
sns.lineplot(data=tmp, x="n_targets", y="f1", label="F1")
plt.ylabel("Model Quality [in %]")
plt.xlabel("Number of Predicted Classes")
plt.savefig(path[:-1] + "_nclasses.pdf")
```
